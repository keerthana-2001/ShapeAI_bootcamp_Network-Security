```python
# Adding salting and iterations to hashes
import hashlib
import uuid
def hash_p(password):
  salt = uuid.uuid4().hex
  s = hashlib.sha256(salt.encode() + password.encode()).hexdigest() + ':' + salt
  return s

def check(p,oldp):
  password, salt = p.split(':')
  a = password == hashlib.sha256(salt.encode() + oldp.encode()).hexdigest()
  return a

password = input("Enter a password: ")
p = hash_p(password)
print("The string to be stored: "+ p)
oldp = input("Enter the old password: ")
if check(p,oldp):
  print("Password matched")
else:
  print("Password does not match")
```

```python
# Python program to generate hashes of string data using 3 algorithms from hashlib
import hashlib
print(hashlib.algorithms_available)
# using sha1 algorithm
data1=hashlib.sha1(b'ShapeAI bootcamps')
print("Using sha1 algorithm: ",end="")
print(data1.hexdigest())
# using sha224 algorithm
data2=hashlib.sha224(b'ShapeAI bootcamps')
print("Using sha224 algorithm: ",end="")
print(data2.hexdigest())
# using sha512 algorithm
data3=hashlib.sha512(b'ShapeAI bootcamps')
print("Using sha512 algorithm: ",end="")
print(data3.hexdigest())
```

```python
# Python program to generate MD5 of string data
import hashlib
strdata = "ShapeAI bootcamps"
value = hashlib.md5(strdata.encode())
print("The byte equivalent of hash is: ",end="")
print(value.digest())
print("The hexadecimal equivalent of hash is: ",end="")
print(value.hexdigest())
```
